```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly
import numpy as np
```

```python
# import demographic spreadsheet as an Excel file
df_happiness = pd.read_excel("Neg_Pos_Affect.xls")
# display first 5 rows of demographic data spreadhseet
df_happiness.head()
```

```python

# Initialize figure
fig = go.Figure()

# Add Traces

fig.add_trace(
    go.Scatter(x=df_happiness["Life Ladder"],
               y=df_happiness["Negative affect"],
               mode='markers',
               name='Negative Affect'))

fig.add_trace(
    go.Scatter(x=df_happiness["Life Ladder"],
               y=df_happiness["Positive affect"],
               mode='markers',
               name='Positive Affect'))

# Set title
fig.update_layout(title_text="Affect and Life Ladder", xaxis_title="Life Ladder",
    yaxis_title="Affect")

plotly.io.write_html(fig,"Affect.html", full_html=False)

fig.show()
```
